```python
import numpy as np
import imageio
from matplotlib import pylab as plt
import keras
%matplotlib inline
```

## II. Random Forests for Image Approximation

### Import image

```python
img = imageio.imread('./mona_lisa.jpg')
plt.imshow(img)
```

### Training Set
Build your ‚Äútraining set‚Äù uniformly sample 5,000 random (x,y) coordinate locations.

#### What other preprocessing steps are necessary for random forests inputs? Do you need to perform mean subtraction, standardization, or unit-normalization?

Random Forests are an extension of decision trees and therefore require little data preparation. They don't need rescaling and can handle both  continuous and discrete features. You don't need to perform mean substraction, standarization or unit normalization.

```python
def RandomSampling(image, N):
    y_rand = np.random.choice(img.shape[0], size=N, replace=True)
    x_rand = np.random.choice(img.shape[1], size=N, replace=True)

    img_masked = np.ones(img.shape)
    points = []
    intensities = []

    for i in range(N):
        # get random coordinate (y,x)
        point = (y_rand[i], x_rand[i])
        # get pixel intensity at the correspondng location
        # rescale to lie between 0.0 and 1.0
        intensity = img[point]/255.0
        points.append(point)
        intensities.append(intensity)
        img_masked[point] = intensity
    return points, intensities, img_masked
```

```python
X_train, y_train, img_masked = RandomSampling(img, 5000)
plt.imshow(img_masked)
```

### Testing Set
Build testing set with all possible coordinates on the image

```python
X_test = []
for y in range(img.shape[0]):
    for x in range(img.shape[1]):
        coord = np.array([y,x])
        X_test.append(coord)
```

### Random Forest Implementation

```python
from sklearn.ensemble import RandomForestRegressor
```

#### Create model

```python
model = RandomForestRegressor(n_estimators=1)
```

#### Train model

```python
model.fit(X_train, y_train)
```

#### Make predictions

```python
y_pred = model.predict(X_test).reshape(img.shape)
plt.imshow(y_pred)
```

### Experimentation

### i. Variable depth
Repeat the experiment for a random forest containing a single decision tree, but with depths 1, 2, 3, 5, 10, and 15.

```python
depths = [1, 2, 3, 5, 10, 15]

fig, axs = plt.subplots(1,3,figsize=(15,4))
for i, depth in enumerate(depths):
    model = RandomForestRegressor(n_estimators=1, max_depth = depth)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test).reshape(img.shape)
    ax = plt.subplot(1,6,i+1)
    plt.imshow(y_pred)
    plt.title("Depth " + str(depth))
fig.suptitle("Single decision tree, variable depth", fontsize=16)
plt.show()
```

#### How does depth impact the result? Describe in detail why.

The maximum depth of the tree controls the resolution of the predicted image. The depth controls how many decisions are made on the tree. With a of depth 1, it means there is only one decision and therefore two possible outcomes, thats why we can see two regions on the first image. For the problem of image regression, depth of the tree is related to the resolution. As the depth is increased, there are more decision regions and therefore it is possible to represent the image more accurately. However, more depth comes with more computational complexity. 

### ii. Variable number of trees
Repeat the experiment for a random forest of depth 7,but with number of trees equalto 1, 3, 5, 10, and 100. 

```python
trees = [1, 3, 5, 10, 100]

fig, axs = plt.subplots(1,3,figsize=(15,4))
for i, tree in enumerate(trees):
    model = RandomForestRegressor(n_estimators=tree, max_depth= 7)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test).reshape(img.shape)
    ax = plt.subplot(1,6,i+1)
    plt.imshow(y_pred)
    plt.title("Number of trees: " + str(tree))
fig.suptitle("Depth of 7, variable decision trees", fontsize=16)
plt.show()
```

#### How does the number of trees impact the result? Describe in detail why.

When the trees have high depth, they can quickly overfit the data. Therefore, the decision boundary between the  classes is very non-smooth and blocky and the regions are highly fragmented.
To reduce this high-variance, random forests uses a bagging technique with multiple decision trees. It averages many models trained on random subsets of the data and the result is a smoothed version. As we see, the image with only 1 tree is fragmented, while the image with 100 trees is smooth. However, increasing the number of trees also comes with more computational cost.

### iv. Custom Pruning
Experiment with different pruning strategies of your choice.

```python
trees = [10, 50]
depths = [10, 15]
node_splits = [2, 3, 7]
leaf_splits = [1, 2]

fig, axs = plt.subplots(1,15,figsize=(15,18))
counter = 0
for leaf_split in leaf_splits:
    for node_split in node_splits:
        for depth in depths:
            for tree in trees:
                model = RandomForestRegressor(n_estimators=tree, 
                                              max_depth= depth, 
                                              min_samples_split=node_split,
                                              min_samples_leaf=leaf_split)
                model.fit(X_train, y_train)
                y_pred = model.predict(X_test).reshape(img.shape)
                counter+=1
                ax = plt.subplot(4, 6,counter)
                plt.axis('off')
                plt.imshow(y_pred)
                plt.title("Trees: "+ str(tree) + 
                          "\n Depth: " + str(depth) + 
                          "\n Node split: " + str(node_split) +
                          "\n Leaf split: " + str(leaf_split))
        
plt.show()
```
